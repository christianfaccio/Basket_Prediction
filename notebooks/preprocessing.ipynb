```python
teams_id = games['HOME_TEAM_ID'].unique()
seasons =  games['SEASON'].unique()
```

```python
games['AWAY_TEAM_WINS'] = games['HOME_TEAM_WINS'].apply(lambda x: 1 if x == 0 else 0)
```

```python
games
```

```python
# Grouping the original data to calculate sums for each TEAM_ID and SEASON_ID
grouped_data_home = games.groupby(['HOME_TEAM_ID', 'SEASON']).agg({'PTS_home':'sum', 
                                                              'FG_PCT_home': 'sum',
                                                                'FT_PCT_home': 'sum',
                                                                'FG3_PCT_home': 'sum',  
                                                                'AST_home': 'sum', 
                                                                'REB_home': 'sum',
                                                              'HOME_TEAM_WINS': 'sum'}).reset_index()
grouped_data_home['HOME_GAMES_PLAYED'] = games.groupby(['HOME_TEAM_ID', 'SEASON']).agg({'HOME_TEAM_WINS':'count'}).reset_index()['HOME_TEAM_WINS']
grouped_data_away = games.groupby(['VISITOR_TEAM_ID', 'SEASON']).agg({'PTS_away':'sum',
                                                                   'FG_PCT_away': 'sum',
                                                                    'FT_PCT_away': 'sum',
                                                                    'FG3_PCT_away': 'sum',
                                                                    'AST_away': 'sum',
                                                                    'REB_away': 'sum',
                                                                    'AWAY_TEAM_WINS': 'sum'}).reset_index()
grouped_data_away['AWAY_GAMES_PLAYED'] = games.groupby(['VISITOR_TEAM_ID', 'SEASON']).agg({'AWAY_TEAM_WINS':'count'}).reset_index()['AWAY_TEAM_WINS']
grouped_data = pd.merge(grouped_data_home, grouped_data_away, left_on=['HOME_TEAM_ID', 'SEASON'], right_on=['VISITOR_TEAM_ID', 'SEASON'], how='outer')
grouped_data.drop(columns=['VISITOR_TEAM_ID'], inplace=True)
grouped_data.rename(columns={'HOME_TEAM_ID': 'TEAM_ID'}, inplace=True)
# ordering by season
grouped_data = grouped_data.sort_values(by=['SEASON', 'TEAM_ID']).reset_index(drop=True)

# y 
grouped_data['win%'] = (grouped_data['HOME_TEAM_WINS'] + grouped_data['AWAY_TEAM_WINS']) / (grouped_data['HOME_GAMES_PLAYED'] + grouped_data['AWAY_GAMES_PLAYED'])

grouped_data.drop(columns=['HOME_TEAM_WINS', 'AWAY_TEAM_WINS', 'HOME_GAMES_PLAYED', 'AWAY_GAMES_PLAYED'], inplace=True)
```

```python
grouped_data
```

```python
# shifting the value to be predicted (win%) by one row behind such that the model can predict the win% for the next season
grouped_data['win%'] = grouped_data.groupby('TEAM_ID')['win%'].shift(-1)
grouped_data.dropna(inplace=True)
```
